A* : f(n) = g(n) + h(n)

Wherein:
g(n) = actual distance from start
h(n) = estimated distance to goal
f(n) = total estimated path cost

```python
#road graph
graph = {
    "Angeles": {"Clark":15, "Dau":30},
    "Clark": {"San Fernando":30},
    "Dau": {"San Fernando": 20},
    "San Fernando" : {"Manila":40},
    "Manila": {}
}
```

```python
import heapq #priority queue

def astar(graph, h, start, goal):
    pq = [(0, start)] # pq(priorityQ)=>(f_cost, node) => "90": Angeles"
    g = {start: 0}# tracking real cost => g(h)
    parent = {} #store each node travel

    while pq: #run=>while pq !empty
        _, node = heapq.heappop(pq) #get=>best node | _ (ignore f_cost)

        if node == goal:
            break

        for n, cost in graph[node].items(): # n=> neigbor node
            new_g = g[node] + cost #new dis. value
            if n not in g or new_g < g[n]: #update
                g[n] = new_g #store best distance
                parent[n] = node #record node traversal
                heapq.heappush(pq, (new_g + h[n], n)) #add to que + compute f(n) = g(n) + h(n) + push calcuated
               # g = 15 / h = 75 = 90 => (90, "Clark") => push to the pq

        path = [goal] #if manila =>strt build path frm goal
        while path[-1] != start: #run == start
                path.append(parent[path[-1]]) #add explored node
                return path[::-1] # [::-1] => reverse list

print("Path:", " -> ".join(path))
print("Cost:", cost)
```
