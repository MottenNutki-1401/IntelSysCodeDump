```python
#graph dictionary
graph = {
    "A": ["B", "C"],
    "B": ['D', 'E'],
    "C": ['F'],
    "D": [],
    "E" : ["F"],
    "F": []
}
```

```python

from collections import deque #double ended queue
def bfs (graph , start) : #bfs => function name (graph, start) => parameters => vars that receives a value
    visited = set() #set (built-in keyword) keep track of nodes we visited, prevent loops

    queue = deque ([start]) #queue => variable for waiting, dequeue=> will act as container [start]=list containing starting node
    visited.add(start) #prevet revisiting later

    while queue:
            node = queue.popleft()  # take the first node from the queue (FIFO rule)
                                   # this is the current node we are exploring
            print (node)

            for neighbors in graph [node]:
                if neighbors not in visited:
                    visited.add(neighbors) #mark as visited
                    queue.append(neighbors)



bfs (graph, "A")
    
```
